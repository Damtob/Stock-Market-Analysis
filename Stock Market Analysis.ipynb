```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
def fetch_stock_data(symbol, start_date, end_date):
    # Fetch stock data
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data
```

```python
def perform_analysis(stock_data):
    # Calculate basic analysis metrics
    max_price = stock_data['High'].max()
    min_price = stock_data['Low'].min()
    avg_price = stock_data['Close'].mean()
    total_volume = stock_data['Volume'].sum()

    print("Stock Analysis:")
    print(f"Maximum Price: {max_price:.3f}")
    print(f"Minimum Price: {min_price:.3f}")
    print(f"Average Closing Price: {avg_price:.3f}")
    print(f"Total Volume: {total_volume:.0f}")
    
   
```

```python
if __name__ == "__main__":
    symbol = "AAPL"  # Replace with the stock symbol you want to analyze; AAPL, GOOGL etc
    start_date = "2023-01-01"
    end_date = "2023-08-11"

    stock_data = fetch_stock_data(symbol, start_date, end_date)
    perform_analysis(stock_data)
```
